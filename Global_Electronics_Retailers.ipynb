**RFM Analysis (Customer Segmentation)**

1. Create Snapshot Date

```python
# We need a reference date (latest order date)
snapshot_date = master_df['Order Date'].max() + pd.Timedelta(days=1)

snapshot_date
```

2. Build RFM Table

```python
rfm = master_df.groupby('CustomerKey').agg({
    'Order Date': lambda x: (snapshot_date - x.max()).days,  # Recency
    'Order Number': 'count',                                # Frequency
    'Revenue USD': 'sum'                                    # Monetary
})

rfm.rename(columns={
    'Order Date': 'Recency',
    'Order Number': 'Frequency',
    'Revenue USD': 'Monetary'
}, inplace=True)

rfm.head()
```

3. Create RFM Scores (1â5 Scale)

```python
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

print(rfm['R_score'].head(5))
print(rfm['F_score'].head(5))
print(rfm['M_score'].head(5))
```

```python
# Combine Scores
rfm['RFM_Score'] = rfm['R_score'].astype(str) + \
                   rfm['F_score'].astype(str) + \
                   rfm['M_score'].astype(str)

print(rfm['RFM_Score'].head(5))
```

4. Customer Segmentation

```python
def segment_customer(row):
    if row['RFM_Score'] == '555':
        return 'Champions'
    elif row['R_score'] >= 4 and row['F_score'] >= 4:
        return 'Loyal Customers'
    elif row['R_score'] >= 4:
        return 'Recent Customers'
    elif row['F_score'] >= 4:
        return 'Frequent Customers'
    elif row['M_score'] >= 4:
        return 'Big Spenders'
    else:
        return 'At Risk'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

print(rfm['Segment'].head(10))
```

5. Segment Distribution

```python
print(rfm['Segment'].value_counts())
```

6. Revenue by Segment

```python
rfm.merge(master_df[['CustomerKey','Revenue USD']], on='CustomerKey') \
   .groupby('Segment')['Revenue USD'].sum() \
   .sort_values(ascending=False)
```

**Product Analysis**

1. Top Selling Products

```python
top_products = master_df.groupby('Product Name').agg({
    'Revenue USD': 'sum',
    'Quantity': 'sum'
}).sort_values(by='Revenue USD', ascending=False).head(10)

top_products
```

2. Category Performance

```python
category_perf = master_df.groupby('Category').agg({
    'Revenue USD': 'sum',
    'Profit USD': 'sum'
}).sort_values(by='Revenue USD', ascending=False)

# Profit Margin
category_perf['Profit Margin %'] = (category_perf['Profit USD'] /
                                    category_perf['Revenue USD']) * 100
category_perf
```

3. Loss-Making Products
